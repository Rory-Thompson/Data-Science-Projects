First we must import the required documentation


```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

### Case 1: M< 1

```python
x = np.linspace(0,20, 300)
y = np.linspace(0,20, 300)
##value for m and c
m = 0.7
c = 2
y = m*x+c##compute the values of y
plt.figure(figsize=(13, 8))
plt.plot(x,y)#plot the function
plt.plot(x,x)#plot y = x
n0= 18
start = 0
plt.legend(["System", "y = x"])
for i in range(10):
    plt.plot((n0,n0), (start,m*n0+c), color = "black")
    
    plt.plot((n0,m*n0+c), (m*n0+c,m*n0+c),color = "black")
    start = m*n0+c
    n0 = m*n0+c
start = 0
n0= 2
for i in range(10):
    plt.plot((n0,n0), (start,m*n0+c), color = "green")
    
    plt.plot((n0,m*n0+c), (m*n0+c,m*n0+c),color = "green")
    start = m*n0+c
    n0 = m*n0+c
plt.title("M < 1")
```

For this case we can see that it does appear to converge to the steady state for when m<1. it seems it would work for all values of the starting value, as the green starting at 2 goes towards the steady state and so does the black values starting at 18.

### Case 2: m>1

```python
x = np.linspace(0,20, 300)
y = np.linspace(0,20, 300)
##value for m and c
m = 2
c = 0
y = m*x+c##compute the values of y
plt.figure(figsize=(13, 8))
plt.plot(x,y)#plot the function
plt.plot(x,x)#plot y = x
n0= 2.5
start = 0
plt.legend(["System", "y = x"])
for i in range(3):
    plt.plot((n0,n0), (start,m*n0+c), color = "black")
    
    plt.plot((n0,m*n0+c), (m*n0+c,m*n0+c), color="black")
    start = m*n0+c
    n0 = m*n0+c
plt.title("M > 1")
```

We can see that from the starting point 2.5, it continues to grow larger and larger and would clearly go onwards to diverge. there still is a chance to get to the steady state but only at 1 intersection point y=x=2x = 0

### Case 3: m = 1 and C = 0

Here we seee that at any starting value it will reach the steady state IE when the function is y = x.

### Case 4: M = 1 c!= 0

```python
x = np.linspace(0,20, 300)
y = np.linspace(0,20, 300)
##value for m and c
m = 1
c = 3
y = m*x+c##compute the values of y
plt.figure(figsize=(13, 8))
plt.plot(x,y)#plot the function
plt.plot(x,x)#plot y = x
n0= 2.5
start = 0
plt.legend(["System", "y = x"])
for i in range(6):
    plt.plot((n0,n0), (start,m*n0+c), color = "black")
    
    plt.plot((n0,m*n0+c), (m*n0+c,m*n0+c), color="black")
    start = m*n0+c
    n0 = m*n0+c
plt.title("M = 1, c != 0")
```

In this case we can see that the function will never reach a steady state as the gradient of the function is 1, the same as y = x, and because c cannot be 0 there will be no intersection point, hence it will continue to oscilate forever.

### Case 5: m = -1

```python
x = np.linspace(0,20, 300)
y = np.linspace(0,20, 300)
##value for m and c
m = -1
c = 10
y = m*x+c##compute the values of y
plt.figure(figsize=(13, 8))
plt.plot(x,y)#plot the function
plt.plot(x,x)#plot y = x
n0= 10
start = 0
plt.legend(["System", "y = x"])
for i in range(6):
    plt.plot((n0,n0), (start,m*n0+c), color = "black")
    
    plt.plot((n0,m*n0+c), (m*n0+c,m*n0+c), color="black")
    start = m*n0+c
    n0 = m*n0+c

n0= 5
start = 0
plt.plot((n0,n0), (start,m*n0+c), color = "green")
plt.title("M = -1")
```

What we see here is that it continues to oscilate around the steady state, not actually converging to the starting value if the intitial value does not equal the intersection point itself.
